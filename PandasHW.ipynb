```python
import math
mean = data.Age.sum() / len(data)
data.Age = data.Age.fillna(mean)
```

## 3

```python
d = {}
for i in range(len(data)):
    name = data.Name[i].split()[1]
    if name in d:
        d[name] += 1
    else:
        d[name] = 1
print(d)
```

## 2

```python
d1 = {}
d2 = {}
d = {}
for i in range(len(data)):
    age = int(data.Age[i]) // 5
    if age in d:
        d[age] += data.Survived[i]
    else:
        d[age] = data.Survived[i]
        
for i in d.keys():
    diap = str(i * 5) + ' - ' + str(i * 5 + 4)
    d1[diap] = d[i]
print(d1)

for i in range(len(data)):
    clas = data.Pclass[i]
    if clas in d2:
        d2[clas] += data.Survived[i]
    else:
        d2[clas] = data.Survived[i]
print('\n', d2)
```

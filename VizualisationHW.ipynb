```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('train.csv')
```

## 1

```python
plt.bar(data.Embarked.value_counts().index, data.Embarked.value_counts().values)
```

## 2

```python
labels = 'men dead', 'women dead', 'women survived', 'men survived'
ms = 0
ws = 0
for i in range(len(data)):
    if data.Sex[i] == 'male':
        ms += data.Survived[i]
    else:
        ws += data.Survived[i]
print(ws + 890 - ws, ms + 890 - ms)
sizes = [pd.value_counts(data.Sex)[0] - ms, pd.value_counts(data.Sex)[1] - ws, ws, ms]
colors = ['yellow', 'orange', 'red', 'darkred']

plt.pie(sizes, labels=labels, colors=colors, shadow=True, autopct='%1.1f%%')
plt.show()
```

## 3

```python
fare = []
for i in range(len(data)):
    fare.append(data.Fare[i])
plt.hist(fare, range(len(data)))
plt.axis([0, len(data), 0, 80])
```

## 4

```python
mean = data.Age.sum() / len(data)
data.Age = data.Age.fillna(mean)
d1 = {}
d = {}
meanf = data.Fare.sum() / len(data)
mass = [meanf] * len(data)
for i in range(len(data)):
    age = int(data.Age[i]) // 5
    if age in d:
        d[age] += 1
    else:
        d[age] = 1
        
for i in d.keys():
    diap = str(i * 5) + '-' + str(i * 5 + 4)
    d1[diap] = d[i]
```

```python
plt.figure()
plt.subplot(211)
plt.bar(d1.keys(), d1.values())


plt.subplot(212)
plt.plot(mass)
```

## 5

```python
d = {'Miss.' : 0, 'Mrs.' : 0, 'Ms.' : 0}
for i in range(len(data)):
    name = data.Name[i].split()[1]
    if name in d:
        d[name] += 1
plt.barh(['Miss', 'Mrs', 'Ms'], d.values())
```
